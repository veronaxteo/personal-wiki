# Probability Distributions

Visualizations of various discrete and continuous distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

## Discrete Distributions

### Binomial

```python
# set parameters
n = 10
p = 0.2

# check for valid parameters
assert n > 0 and isinstance(n, int) and 0 < p < 1, 'Parameters are invalid.'

# binomial distribution
x = st.binom(n, p)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pmf)
ax = fig.add_subplot(121)
ax.set_title('binomial pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, n, n+1)                          
plt.plot(val, x.pmf(val), ls='None', marker='o', markersize=5)      

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('binomial cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, n + 1, 1000)
plt.plot(val, x.cdf(val));
```

### Poisson

```python
# set parameter
lambd = 4

# poisson distribution
x = st.poisson(lambd)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pmf)
ax = fig.add_subplot(121)
ax.set_title('poisson pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, 14, 15)                          
plt.plot(val, x.pmf(val), ls='None', marker='o', markersize=5) 

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('poisson cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, 15, 1000)
plt.plot(val, x.cdf(val));
```

### Hypergeometric

```python
# set parameters
n = 10
M = 20
N = 50

# hypergeometric distribution
x = st.hypergeom(N, n, M)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pmf)
ax = fig.add_subplot(121)
ax.set_title('hypergeometric pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, 10, 11)
plt.plot(val, x.pmf(val), ls='None', marker='o')

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('hypergeometric cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, 11, 1000)
plt.plot(val, x.cdf(val));
```

### Geometric

```python
# set parameters
r = 4
p = 0.45

# geometric (negative binomial)
x = st.nbinom(r, p)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pmf)
ax = fig.add_subplot(121)
ax.set_title('pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, 20, 21)                         
plt.plot(val, x.pmf(val), ls='None', marker='o', markersize=5)     

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, 21, 1000)                 
plt.plot(val, x.cdf(val));
```

## Continuous Distributions

### Normal

```python
# set parameters 
mu = 10
sigma = 4

# normal distribution
x = st.norm(loc=mu, scale=sigma)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pdf)
ax = fig.add_subplot(121)
ax.set_title('normal pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
val = np.linspace(-5, 25, 1000)
plt.plot(val, x.pdf(val))            

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('normal cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-5, 25, 1000)
plt.plot(val, x.cdf(val));
```

### Uniform

```python
# set parameters 
a = 8
b = 22

# uniform distribution
x = st.uniform(loc=a, scale=b-a)

# figura que tendrá los dos gráficos
fig = plt.figure(figsize=(12, 4))

# graph of disttribution (pdf)
ax = fig.add_subplot(121)
ax.set_title('uniform pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
val = np.linspace(0, 30, 1000)
plt.plot(val, x.pdf(val))

# graph of disttribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('uniform cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(0, 30, 1000)
plt.plot(val, x.cdf(val));
```

### Exponential

```python
# set parameters
lambd = 0.2

# exponential distribution
x = st.expon(scale=1/lambd, loc=0)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pdf)
ax = fig.add_subplot(121)
ax.set_title('exponential pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
val = np.linspace(0, 40, 1000)
ax.plot(val, x.pdf(val))

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('exponential cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(0, 40, 1000)
ax.plot(val, x.cdf(val));
```

### Triangular

```python
# set parameters
a = 5
b = 25
c = 10

# triangular distribution
x = st.triang((c-a)/(b-a), loc=a, scale=b-a)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pdf)
ax = fig.add_subplot(121)
ax.set_title('triangular pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
val = np.linspace(0, 30, 1000)
plt.plot(val, x.pdf(val))

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('triangular cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(0, 30, 1000)
plt.plot(val, x.cdf(val));
```

### Trapezoidal

```python
# set parameters
a = 2
b = 25
m1 = 6
m2 = 15

# trapezoidal distribution
x = st.trapz((m1-a)/(b-a), (m2-a)/(b-a), loc=a, scale=b-a)

# figura que tendrá los dos gráficos
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pdf)
ax = fig.add_subplot(121)
ax.set_title('trapezoidal pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
val = np.linspace(0, 30, 1000)
plt.plot(val, x.pdf(val))

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('trapezoidal cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(0, 30, 1000)
plt.plot(val, x.cdf(val));
```

### Gamma

```python
# set parameters
a = 4
b = 2.5

# gamma distribution
x = st.gamma(a, loc=0, scale=b)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pdf)
ax = fig.add_subplot(121)
ax.set_title('gamma pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
val = np.linspace(0, 40, 1000)
plt.plot(val, x.pdf(val))

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('gamma cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(0, 40, 1000)
plt.plot(val, x.cdf(val));
```

### Beta

```python
# set parameters
a = -1
b = 5
alfa = 0.8
beta = 0.8

# check for valid parameters
assert b > a and alfa > 0 and beta > 0, 'Parameters are invalid.'

# beta distribution
x = st.beta(alfa, beta, loc=a, scale=b-a)

# set figure size
fig = plt.figure(figsize=(12, 4))

# graph of distribution (pdf)
ax = fig.add_subplot(121)
ax.set_title('beta pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
val = np.linspace(a, b, 1000)
plt.plot(val, x.pdf(val))

# graph of distribution (cdf)
ax = fig.add_subplot(122)
ax.set_title('beta cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(a, b, 1000)
plt.plot(val, x.cdf(val));
```
